```python
# import pandas and matplotlib modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("alphabet_stock_data.csv")
```

```python
data
```

```python
start_date = pd.to_datetime('2020-05-01')
end_date = pd.to_datetime('2020-09-19')
data['Date'] = pd.to_datetime(data['Date'])
```

```python
within_span = (data['Date'] >= start_date) & (data['Date'] <= end_date)
```

```python
fix = data[within_span]
```

```python
fix
```

```python
fix_index = fix.set_index('Date')
fix_index
```

```python
plt.figure(figsize=(10,10))
fix_index['Close'].plot(color='red')
plt.title('Stock Prices of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('$ Price')
```
